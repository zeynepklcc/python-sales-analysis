```python

sınıf_mevcudu=int(input("Toplam Öğrenci Sayısını Giriniz: "))


vize_notları=[]
final_notları=[]
genel_notlar=[]
harf_notları=[]


i=1

while i<=sınıf_mevcudu:
    vize_notu=int(input(f"{i}. Öğrencinin Vize Notunu Giriniz: "))
    if vize_notu>100 or vize_notu<0:
        print("Lütfen 0 ve 100 arasında bir değer giriniz")
        continue
    vize_notları.append(vize_notu)
    final_notu=int(input(f"{i}. Öğrencinin Final Notunu Giriniz: "))
    if final_notu>100 or final_notu<0:
         print("Lütfen 0 ve 100 arasında bir değer giriniz")
         continue
    final_notları.append(final_notu)

    genel_not=vize_notu*0.4+final_notu*0.6
    genel_notlar.append(genel_not)


    if genel_not>=85:
        harf_notu="AA"
    elif genel_not>=75:
        harf_notu="BA"
    elif genel_not>=65:
        harf_notu="BB"
    elif genel_not>=55:
        harf_notu="CB"
    elif genel_not>=50:
        harf_notu="CC"
    elif genel_not>=45:
        harf_notu="DC"
    elif genel_not>=40:
        harf_notu="DD"
    else:
        harf_notu="FF"

    harf_notları.append(harf_notu)

    i+=1
    continue


print(f"Vize Notları: {vize_notları}")
print(f"Final Notları: {final_notları}")
print(f"\nGenel Notlar: {genel_notlar}")
print(f"Harf Notları: {harf_notları}")


vize_sınıf_ortalaması=sum(vize_notları)/sınıf_mevcudu
final_sınıf_ortalaması=sum(final_notları)/sınıf_mevcudu
sınıf_ortalaması=sum(genel_notlar)/sınıf_mevcudu


print(f"\nVize Notlarının Ortalaması: {vize_sınıf_ortalaması}")
print(f"Final Notlarının Ortalaması: {final_sınıf_ortalaması}")
print(f"Sınıfın Ortalaması: {sınıf_ortalaması:.1f}")


for j in range(0,sınıf_mevcudu+1):
    vize_notu=vize_notları[j]
    final_notu=final_notları[j]
    harf_notu=harf_notları[j]
    print(f"{j+1}. Öğrencinin Sınav Bilgileri: {vize_notu} - {final_notu} - {harf_notu}")
    continue


AA_harf_notu=harf_notları.count("AA")
BA_harf_notu=harf_notları.count("BA")
BB_harf_notu=harf_notları.count("BB")
CB_harf_notu=harf_notları.count("CB")
CC_harf_notu=harf_notları.count("CC")
DC_harf_notu=harf_notları.count("DC")
DD_harf_notu=harf_notları.count("DD")
FF_harf_notu=harf_notları.count("FF")


print("\nHarf Notları: ")
print(f"AA Harf Notu: %{AA_harf_notu*100/sınıf_mevcudu:.0f}")
print(f"BA Harf Notu: %{BA_harf_notu*100/sınıf_mevcudu:.0f}")
print(f"BB Harf Notu: %{BB_harf_notu*100/sınıf_mevcudu:.0f}")
print(f"CB Harf Notu: %{CB_harf_notu*100/sınıf_mevcudu:.0f}")
print(f"CC Harf Notu: %{CC_harf_notu*100/sınıf_mevcudu:.0f}")
print(f"DC Harf Notu: %{DC_harf_notu*100/sınıf_mevcudu:.0f}")
print(f"DD Harf Notu: %{DD_harf_notu*100/sınıf_mevcudu:.0f}")
print(f"FF Harf Notu: %{FF_harf_notu*100/sınıf_mevcudu:.0f}\n")


print(f"****Çok Başarılı Öğrenci Sayısı: {AA_harf_notu+BA_harf_notu+BB_harf_notu}")
print(f"**Başarılı Öğrenci Sayısı: {CB_harf_notu+CC_harf_notu+BB_harf_notu}")
print(f"*Orta Seviye Başarılı Öğrenci Sayısı: {DC_harf_notu+BA_harf_notu+DD_harf_notu}")
print(f"Başarısız Öğrenci Sayısı: {FF_harf_notu}")




```
