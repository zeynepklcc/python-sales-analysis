```python

sınıf_mevcudu=int(input("Toplam Öğrenci Sayısını Giriniz: "))


vize_notları=[]
final_notları=[]
ortalama_notlar=[]
harf_notları=[]


i=1

while i<=sınıf_mevcudu:
    vize_notu=int(input(f"{i}. Öğrencinin Vize Notunu Giriniz: "))
    if vize_notu>100 or vize_notu<0:
        print("Lütfen 0 ve 100 arasında bir değer giriniz")
        continue
    vize_notları.append(vize_notu)
    final_notu=int(input(f"{i}. Öğrencinin Final Notunu Giriniz: "))
    if final_notu>100 or final_notu<0:
         print("Lütfen 0 ve 100 arasında bir değer giriniz")
         continue
    final_notları.append(final_notu)
    def not_hesapla(vize_notu,final_notu):    
        genel_not=vize_notu*0.4+final_notu*0.6
        harf_notu_eşik=[(85,"AA"),(75,"BA"),(65,"BB"),(55,"CB"),(45,"CC"),(40,"DC"),(35,"DD"),(0,"FF")]
        for esik,harf in harf_notu_eşik:
            if genel_not>=esik:
                harf_notu=harf
                harf_notları.append(harf_notu)
                ortalama_notlar.append(genel_not)
                print(f"\nÖğrencinin Ortalaması: {genel_not}\nÖğrencinin Harf Notu: {harf_notu}")
                print(f"\nGenel Ortalama: {sum(ortalama_notlar)/sınıf_mevcudu:.1f}\n")
                break    
    not_hesapla(vize_notu,final_notu)
    i+=1

print("Harf Notu Yüzdelikleri: ")
harf_notları2=set(harf_notları)
for x in harf_notları2:
    result=harf_notları.count(x)
    print(f"{x} Harf Notu: %{result*100/sınıf_mevcudu}")

```
